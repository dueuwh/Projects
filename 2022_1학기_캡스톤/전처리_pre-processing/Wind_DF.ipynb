```python
from google.colab import drive
drive.mount('/content/drive')
import numpy as np
import pandas as pd
import datetime
import matplotlib.pyplot as plt
import random
import lightgbm as lgb

train_input = pd.read_csv("/content/drive/MyDrive/발전예측/2ndSpline/wind_train_data.csv", index_col=0)
train_input.drop(["date"], axis=1, inplace=True)

plus_data = pd.read_csv("/content/drive/MyDrive/발전예측/2ndSpline/df_2020_to_2021_v2_rain_flag_pulse", index_col=0)
plus_data = plus_data.drop(["sunlight", "insolation", "datetime", "date"], axis=1)

train_input = pd.concat([train_input, plus_data], axis=1)

index_0 = int(train_input.shape[0])
index_1 = int(train_input.shape[0]*0.85)
index_2 = int(train_input.shape[0]*0.95)
train_slice = slice(0,index_1)
valid_slice = slice(index_1, index_2)
test_slice = slice(index_2, index_0)

train_ = train_input[train_slice]
train_ = pd.concat([train_, train_], axis=0)
valid_ = train_input[valid_slice]
valid_ = pd.concat([valid_, valid_], axis=0)
test_ = train_input[test_slice]

train_x = train_.drop(["wind_speed"], axis=1)
train_y = train_["wind_force"]
valid_x = valid_.drop(["wind_speed"], axis=1)
valid_y = valid_["wind_force"]
test_x = test_.drop(["wind_speed"], axis=1)
test_y = test_["wind_force"]

valid_x.reset_index(drop=True, inplace=True)
valid_y.reset_index(drop=True, inplace=True)
test_x.reset_index(drop=True, inplace=True)
test_y.reset_index(drop=True, inplace=True)

train_y = list(train_y)
valid_y = list(valid_y)
test_y = list(test_y)
```

```python
wind_model_01 = lgb.LGBMRegressor(learning_rate=0.08353774569575682, max_bin=1023,
                                  objective='quantile', metric=['quantile'], alpha=0.1,
                                  min_child_samples=3, n_estimators=88, num_leaves=4,
                                  reg_alpha=0.0009765625, reg_lambda=0.0011091735927831995,
                                  n_iter=10000000, verbose=-1)
```

```python
wind_model_05 = lgb.LGBMRegressor(learning_rate=0.08353774569575682, max_bin=1023,
                                  objective='quantile', metric=['quantile'], alpha=0.5,
                                  min_child_samples=3, n_estimators=88, num_leaves=4,
                                  reg_alpha=0.0009765625, reg_lambda=0.0011091735927831995,
                                  n_iter=10000000, verbose=-1)
```

```python
wind_model_09 = lgb.LGBMRegressor(learning_rate=0.08353774569575682, max_bin=1023,
                                  objective='quantile', metric=['quantile'], alpha=0.9,
                                  min_child_samples=3, n_estimators=88, num_leaves=4,
                                  reg_alpha=0.0009765625, reg_lambda=0.0011091735927831995,
                                  n_iter=10000000, verbose=-1)
```

```python
wind_model_01.fit(train_x, train_y,
              eval_set=[(valid_x, valid_y)],
              eval_metric='l1',
              early_stopping_rounds=1000
            )
```

```python
wind_model_01.booster_.save_model('wind_model_01.txt')
```

```python
wind_model_09.fit(train_x, train_y,
              eval_set=[(valid_x, valid_y)],
              eval_metric='l1',
              early_stopping_rounds=1000
            )
```

```python
wind_model_09.booster_.save_model('wind_model_09.txt')
```

```python
wind_model_05.fit(train_x, train_y,
              eval_set=[(valid_x, valid_y)],
              eval_metric='l1',
              early_stopping_rounds=1000
            )
wind_model_05.booster_.save_model('wind_model_05.txt')
```

```python
wind_01_predict = wind_model_01.predict(test_x)
wind_05_predict = wind_model_05.predict(test_x)
wind_09_predict = wind_model_09.predict(test_x)
```

```python
show_location = 500
plt.figure(figsize=(20,5))
fill_range = np.arange(0,len(wind_01_predict))
plt.plot(wind_01_predict, c = 'b', label = 'quantile_0.1')
plt.plot(wind_05_predict, c = 'C1', label = 'quantile_0.5')
plt.plot(wind_09_predict, c = 'y', label = 'quantile_0.9')
plt.fill_between(fill_range, wind_01_predict[:], wind_09_predict[:], color='lightgray')
plt.plot(test_y, c = 'r', label = 'True')
plt.title('quantile 0.1 & 0.5 & 0.9, gray : 0.1 ~ 0.9')
plt.xlim(show_location, show_location+600)
plt.legend()
plt.show()
```
